```python
import pandas as pd 
population = pd.read_csv('population.csv', index_col='age group')
print(population)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

population = pd.read_csv('population.csv',index_col='age group')

plt.figure(figsize=(15, 8))

bar_groups = ['0-14', '15-64', 'above 65']
bar_width = 0.25
index = np.arange(len(population.columns))
for i, row in enumerate(bar_groups):
    plt.bar(index + i * bar_width, population.loc[row], bar_width, label=row)

for row in population.index:
    if row not in bar_groups: 
        plt.plot(population.columns, population.loc[row], marker='o', label=row)
        for x, y in zip(population.columns, population.loc[row]):
            plt.text(x, y, f'{y:,}', ha='center', va='top', fontsize=10)



plt.title('Population (2014-2023)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Population')

plt.legend(title="Age Group")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

birthrate = pd.read_csv('birthrate.csv',index_col='year')

plt.figure(figsize=(15, 8))
plt.plot(birthrate.index, birthrate['birthrate'], marker='o', label='Birthrate') 

for x, y in zip(birthrate.index, birthrate['birthrate']):
    plt.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=13)

plt.title('Birthrate (2014-2023)')
plt.xlabel('Year')
plt.ylabel('rate %')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

unemployment_rate = pd.read_csv('unemployment rate.csv',index_col='year')

plt.figure(figsize=(15, 8))

plt.plot(unemployment_rate.index, unemployment_rate['unemployment rate(%)'], marker='o', label='Unemployment Rate')
plt.plot(unemployment_rate.index, unemployment_rate['birthrate'], marker='o', label='Birthrate')

for x, y in zip(unemployment_rate.index, unemployment_rate['birthrate']):
    plt.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=13)
for x, y in zip(unemployment_rate.index, unemployment_rate['unemployment rate(%)']):
    plt.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=13)

plt.title('Unemployment Rate and Birthrate (2014-2023)')
plt.xlabel('Year')
plt.ylabel('rate %')

plt.legend()

plt.grid(True)


plt.show()
```
